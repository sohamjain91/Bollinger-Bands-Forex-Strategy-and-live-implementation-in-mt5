```python
def bb_fx(df, bb_period, bb_std, rolling_high, rr, capital, risk_per_trade):
    
    df['ma'] = df.Close.rolling(bb_period).mean()
    df['std'] = df.Close.rolling(bb_period).std()
    df['upper_bb'] = df['ma'] + 2*df['std']
    df['lower_bb'] = df['ma'] - 2*df['std']
    df['roll_high'] = df['High'].rolling(rolling_high).max()
    df['roll_close'] = df['Close'].rolling(rolling_high).max()
    df = df.dropna()
    
    look = False
    trade = False
    ready = False

    equity_curve = []

    for i in range(len(df)-10):

        try:

            if (look == False) & (trade == False):

                if (df['Close'].iloc[i] > df['upper_bb'].iloc[i]):

                    index_i = i
                    upper_line = df['High'].iloc[i]

                    if (df['Close'].iloc[i+1]<df['upper_bb'].iloc[i+1]) & (df['roll_high'].iloc[i+rolling_high]<upper_line):

                        look = True
                        position = -1

                elif (df['Close'].iloc[i] < df['lower_bb'].iloc[i]):

                    index_i = i
                    upper_line = df['Low'].iloc[i]

                    if (df['Close'].iloc[i+1]>df['lower_bb'].iloc[i+1]) & (df['roll_low'].iloc[i+rolling_high]>upper_line):

                        look = True           
                        position = 1 

            if position == -1:

                if ((look == True) & (trade == False)):

                    if (df['High'].iloc[i+rolling_high+1]>upper_line) & (df['Close'].iloc[i+rolling_high+1]<upper_line):

                        sl = df['High'].iloc[i+rolling_high+1]
                        limit = df['Low'].iloc[i+rolling_high+1]

                        ready = True

                    elif df['Close'].iloc[i+rolling_high+1]>upper_line:

                        look = False

                if ((look == True) & (ready == True) & (trade == False)):

                    if (df['Close'].iloc[i+rolling_high+2]<limit):

                        trade = True
                        entry_price = df['Close'].iloc[i+rolling_high+2]
                        sl_depth = sl - entry_price
                        tp_depth = rr * sl_depth
                        tp = entry_price - tp_depth
                        ready = False

                if (look == True) & (trade == True):

                    if (df['Close'].iloc[i+rolling_high+3]<tp):
                        exit_price = df['Close'].iloc[i+rolling_high+3]
                        look = False

                    if (df['Close'].iloc[i+rolling_high+3]< df['lower_bb'].iloc[i+rolling_high+3]): 
                        exit_price = df['Close'].iloc[i+rolling_high+3]
                        look = False

                    if (df['Close'].iloc[i+rolling_high+3]>sl):
                        exit_price = df['Close'].iloc[i+rolling_high+3]
                        look = False

                if ((look == False) & (trade == True)):

                    profit_per_contract = entry_price - exit_price
                    rpt = capital * risk_per_trade
                    contracts = rpt/sl_depth
                    lots = round(contracts/100000, 2)
                    total_profit = profit_per_contract * contracts
                    capital = capital + total_profit
                    equity_curve.append([df.index[index_i], capital, tp, sl, limit, upper_line, lots, position])

                    trade = False

            if position == 1: 

                if ((look == True) & (trade == False)):

                    if (df['Low'].iloc[i+rolling_high+1]<upper_line) & (df['Close'].iloc[i+rolling_high+1]>upper_line):

                        sl = df['Low'].iloc[i+rolling_high+1]
                        limit = df['High'].iloc[i+rolling_high+1]

                        ready = True

                    elif df['Close'].iloc[i+rolling_high+1]<upper_line:

                        look = False

                if ((look == True) & (ready == True) & (trade == False)):

                    if (df['Close'].iloc[i+rolling_high+2]>limit):

                            trade = True
                            entry_price = df['Close'].iloc[i+rolling_high+2]
                            sl_depth = entry_price - sl
                            tp_depth = rr * sl_depth
                            tp = tp_depth + entry_price
                            ready = False

                if (look == True) & (trade == True):

                    if (df['Close'].iloc[i+rolling_high+3]>tp):
                        exit_price = df['Close'].iloc[i+rolling_high+3]
                        look = False

                    if (df['Close'].iloc[i+rolling_high+3]> df['lower_bb'].iloc[i+rolling_high+3]): 
                        exit_price = df['Close'].iloc[i+rolling_high+3]
                        look = False

                    if (df['Close'].iloc[i+rolling_high+3]<sl):
                        exit_price = df['Close'].iloc[i+rolling_high+3]
                        look = False

                if ((look == False) & (trade == True)):

                    profit_per_contract =  exit_price - entry_price
                    rpt = capital * risk_per_trade
                    contracts = rpt/sl_depth
                    lots = round(contracts/100000, 2)
                    total_profit = profit_per_contract * contracts
                    capital = capital + total_profit
                    equity_curve.append([df.index[index_i], capital, tp, sl, limit, upper_line, lots, position])

                    trade = False
                    
        except Exception as e:

            pass
        
    return(capital)
```
